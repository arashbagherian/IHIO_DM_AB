```python
import pandas as pd
import numpy as np
from statsmodels.stats.proportion import proportion_confint
```

# DM

## Importing

```python
dm_df = pd.read_pickle('./dm_drug_class_numbers_encr.pkl')
dm_df
```

## Building dm drug-class Table

```python
dm_classes = ['A10A', 'A10BA', 'A10BB', 'A10BD', 'A10BF', 'A10BG', 'A10BJ', 'A10BX']
```

```python
dm_table = pd.DataFrame(columns=sorted(dm_df['cat'].unique()))
```

```python
dm_table.loc['total'] = dm_df.groupby('cat')[dm_classes].sum().T.sum()
```

```python
dm_table
```

```python
dm_table = pd.concat(
    [dm_table, dm_df.groupby('cat')[dm_classes].sum().T]
)
```

```python
dm_concat_dict = dict()
```

```python
dm_concat_dict['numbers'] = dm_table
```

```python
dm_concat_dict['proportions'] = dm_table.loc[dm_classes] / dm_table.loc['total']
```

```python
pd.concat(dm_concat_dict, axis=1)
```

```python
dm_concat_dict['ci_low'], dm_concat_dict['ci_up'] = proportion_confint(
    dm_table.loc[dm_classes],
    dm_table.loc['total']
)
```

```python
pd.concat(dm_concat_dict, axis=1)
```

# AB

## Importing

```python
ab_df = pd.read_pickle('./ab_drug_class_numbers_encr.pkl')
ab_df.drop(ab_df.loc[ab_df['cat'] == 0].index, axis=0, inplace=True)
ab_df
```

## Building ab drug-class Table

```python
ab_classes = ['J01A', 'J01B', 'J01C', 'J01D', 'J01E', 'J01F', 'J01G', 'J01M', 'J01X']
```

```python
ab_table = pd.DataFrame(columns=sorted(ab_df['cat'].unique()))
```

```python
ab_table.loc['total_service'] = ab_df.groupby('cat')[ab_classes].sum().T.sum()
```

```python
ab_table.loc['total_person'] = ab_df['cat'].value_counts()
```

```python
ab_table
```

```python
ab_table = pd.concat(
        [ab_table, ab_df.groupby('cat')[ab_classes].sum().T]
)
```

```python
ab_concat_dict = dict()
```

```python
ab_concat_dict['numbers'] = ab_table
```

```python
ab_concat_dict['number/total_service'] = ab_table.loc[ab_classes] / ab_table.loc['total_service']
```

```python
ab_concat_dict['s_ci_low'], ab_concat_dict['s_ci_up'] = proportion_confint(
    ab_table.loc[ab_classes],
    ab_table.loc['total_service']
)
# Changing column names
ab_concat_dict['s_ci_low'].columns = ab_table.columns
ab_concat_dict['s_ci_up'].columns = ab_table.columns
```

```python
ab_concat_dict['number/person'] = (
    ab_table.loc[ab_classes] / ab_table.loc['total_person']
)
```

```python
ab_concat_dict['p_ci_low'], ab_concat_dict['p_ci_up'] = proportion_confint(
    ab_table.loc[ab_classes],
    ab_table.loc['total_person']
)
# Changing column names
ab_concat_dict['p_ci_low'].columns = ab_table.columns
ab_concat_dict['p_ci_up'].columns = ab_table.columns
```

```python
pd.concat(ab_concat_dict,  axis=1)
```

```python
pd.concat(ab_concat_dict, axis=1)
```

```python
temp = pd.DataFrame(columns=sorted(ab_df['cat'].unique()))
for cat in sorted(ab_df['cat'].unique()):
    temp_cat = ab_df.loc[ab_df['cat'] == cat]
    temp[cat] = temp_cat[ab_classes].isna().sum() / temp_cat.shape[0]
```

# Dominant Drug

## Importing

```python
dominant_df = pd.read_pickle('./results/dominants_0.8.pkl')
# By checking the results, we found out we don't need to consider one-time use
# of insulin.
dominant_df = dominant_df[0.8].to_frame('drug')
dominant_df
```

```python
np.ones(dominant_df['drug'].nunique()) * dominant_df['drug'].shape[0]
```

## Building Dominant Table

```python
def my_proportion_confint(df, col):
    low, up = proportion_confint(
        df['drug'].value_counts(),
        df.shape[0]
    )
    return low, up
```

```python
dominant_df['cat'] = dm_df['cat']
```

```python
dominant_df['cat'].value_counts()
```

```python
major_classes = [
    'A10BB',
    'A10A A10BA',
    'A10BA A10BB A10BG',
    'A10BA A10BG',
    'A10BG',
    'A10BA',
    'A10A',
    'A10A A10BA A10BB',
    'A10BA A10BB',
    'A10BA A10BB A10BF'
 ]
```

```python
dominant_df.loc[~ dominant_df['drug'].isin(major_classes), 'drug'] = 'other'
```

```python
dominant_proportions = dominant_df.groupby('cat')['drug'].value_counts(normalize=True)
dominant_proportions
```

```python
dominant_proportions[0].sort_values(ascending=False).to_frame('proportion').index
```

```python
dominant_dict = dict()
```

```python
for cat in sorted(dm_df['cat'].unique()):
    # Proportion
    # This also drops na so I can use the shape for confint
    temp_cat = dominant_proportions[cat].sort_values(ascending=False).to_frame(
        'proportion')
    # confidence intervals
    temp_cat['ci_low'], temp_cat['ci_up'] = my_proportion_confint(
        dominant_df.loc[dominant_df['cat'] == cat, ['drug']],
        'drug'
    )
    # Adding drug as a column
    temp_cat.reset_index(inplace=True)
    # Cumulative proportion sum
    temp_cat['cumulative'] = temp_cat['proportion'].cumsum().values
    dominant_dict[cat] = temp_cat
```

```python
pd.concat(dominant_dict, axis=1)
```
