```python
import pandas as pd
import numpy as np
```

```python
dollor_coef = 25942
```

```python
def build_string_0(val_df, low_df, up_df, to_percentage):

    if to_percentage:
        val_df = val_df * 100
        low_df = low_df * 100
        up_df = up_df * 100

    val_df = val_df.map(lambda x: f'{x:.0f}')
    low_df = low_df.map(lambda x: f'{x:.0f}')
    up_df = up_df.map(lambda x: f'{x:.0f}')



    return val_df + ' (' + low_df + '-' + up_df + ')'


def build_string_2(val_df, low_df, up_df, to_percentage):

    if to_percentage:
        val_df = val_df * 100
        low_df = low_df * 100
        up_df = up_df * 100

    val_df = val_df.map(lambda x: f'{x:.2f}')
    low_df = low_df.map(lambda x: f'{x:.2f}')
    up_df = up_df.map(lambda x: f'{x:.2f}')



    return val_df + ' (' + low_df + '-' + up_df + ')'
```

## Mean CI

```python
# Importing
mean_df = pd.read_excel('../costs_by_variables_mean_ci_f.xlsx',
                        index_col=[0, 1], header=[0, 1, 2])
mean_df = mean_df / dollor_coef
```

```python
mean_df
```

```python
# Removing all insurance columns
mean_df.drop('insurance', axis=1, level=2, inplace=True)

mean_df[('val', 'ab', 'sum')]
```

```python
build_string_0(
    mean_df[('val', 'ab', 'sum')],
    mean_df[('low', 'ab', 'sum')],
    mean_df[('up', 'ab', 'sum')],
    to_percentage=False,
)
```

```python
mean_table = pd.DataFrame(index=mean_df['val'].index, columns=mean_df['val'].columns)
```

```python
mean_df.loc[:, pd.IndexSlice['val', :, 'sum']]
```

```python
# Putting in total
mean_table[('total', 'total')] = build_string_0(
    mean_df[('val', 'total', 'total')],
    mean_df[('low', 'total', 'total')],
    mean_df[('up', 'total', 'total')],
    to_percentage=False
)
```

```python
mean_table
```

```python
# calculating proportions
# pocket / sum
for col1 in ['val', 'low', 'up']:
    for col2 in ['ab', 'dm', 'other']:
        mean_df.loc[:, (col1, col2, 'pocket')] = (
            mean_df.loc[:, (col1, col2, 'pocket')] / mean_df.loc[:, ('val', col2, 'sum')]
            )
# sum / total
for col1 in ['val', 'low', 'up']:
    for col2 in ['ab', 'dm', 'other']:
        mean_df.loc[:, (col1, col2, 'sum')] = (
            mean_df.loc[:, (col1, col2, 'sum')] / mean_df.loc[:, ('val', 'total', 'total')]
            )
```

```python
# Building string and putting it in the table.
for col2 in ['ab', 'dm', 'other']:
    
    # pocket
    mean_table.loc[:, (col2, 'pocket')] = build_string_2(
        mean_df[('val', col2, 'pocket')],
        mean_df[('low', col2, 'pocket')],
        mean_df[('up', col2, 'pocket')],
        to_percentage=True
    )

    # sum
    mean_table.loc[:, (col2, 'sum')] = build_string_2(
        mean_df[('val', col2, 'sum')],
        mean_df[('low', col2, 'sum')],
        mean_df[('up', col2, 'sum')],
        to_percentage=True
    )
```

```python
mean_table
```

### Decoding province names

```python
# province_id
province_id = pd.read_excel('../../diabeteschronicabuserprojectihio/Test_sample/Test/province_id.xlsx')
province_mapper = province_id.set_index('province_id_iso')['province_name_en']
```

```python
def rename_second_level(df, specific_level_0, mapper):

    # Reseting index 0 and 1
    df = df.reset_index([0, 1])

    # return mapper[df.loc[df['level_0'] == specific_level_0, 'level_1']].values

    df.loc[df['level_0'] == specific_level_0, 'level_1'] =\
        mapper[df.loc[df['level_0'] == specific_level_0, 'level_1']].values



    return df.set_index(['level_0', 'level_1'])
```

## Median IQR

```python
median_df = pd.read_excel('../costs_by_variables_quartiles_f.xlsx',
                          index_col=[0, 1], header=[0, 1, 2])
median_df = median_df / dollor_coef
```

```python
# Removing all insurance columns
median_df.drop('insurance', axis=1, level=2, inplace=True)
```

```python
# Computing iqr
iqr = median_df[0.75] - median_df[0.25]
```

```python
median_table = pd.DataFrame(index=iqr.index, columns=iqr.columns)
median_table
```

```python
def build_string_iqr_0(val_df, iqr_df):

    val_df = val_df.map(lambda x: f'{x:.0f}')
    iqr_df = iqr_df.map(lambda x: f'{x:.0f}')

    return val_df + ' (' + iqr_df + ')'

def build_string_iqr_2(val_df, iqr_df):

    val_df = val_df.map(lambda x: f'{x:.2f}')
    iqr_df = iqr_df.map(lambda x: f'{x:.2f}')

    return val_df + ' (' + iqr_df + ')'


def build_string_2(val_df, to_percentage):

    if to_percentage:
        val_df = val_df * 100

    val_df = val_df.map(lambda x: f'{x:.2f}')

    return val_df
```

```python
# Putting in total
median_table[('total', 'total')] = build_string_iqr_2(
    median_df[(0.5, 'total', 'total')],
    iqr[('total', 'total')]
)
```
