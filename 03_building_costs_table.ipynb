## Computing Confidence Intervals

```python
import scipy.stats as st


def get_ci(g):
    low, up = st.t.interval(confidence=0.95, df=g.shape[0]-1,
                            loc=np.mean(g, axis=0),
                            scale=st.sem(g))
    return low, up
```

```python
concat_dict = dict()
for col in data.columns:
    concat_dict[col] = costs.groupby(data[col])[costs.columns].apply(lambda g: get_ci(g))
```
