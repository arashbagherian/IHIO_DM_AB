```python
# Example data
studies = ['Study 1', 'Study 2', 'Study 3', 'Study 4']
odds_ratios = [1.5, 0.75, 1.25, 2.0]
confidence_intervals = [(1.2, 1.8), (0.5, 1.0), (0.9, 1.6), (1.5, 2.5)]

# Plotting
fig, ax = plt.subplots(figsize=(6, 4))

# Horizontal lines for confidence intervals
for i, (ci, orr) in enumerate(zip(confidence_intervals, odds_ratios)):
    ax.plot(ci, [i, i], color='black')  # Confidence interval
    ax.plot(orr, i, 'ro')  # Red dot for odds ratio

# Customize y-axis with study labels
ax.set_yticks(range(len(studies)))
ax.set_yticklabels(studies)

# Add a vertical line at OR=1 (no effect)
ax.axvline(x=1, linestyle='--', color='grey')

# Labels and title
ax.set_xlabel('Odds Ratio')
ax.set_title('Odds Ratio Plot')

plt.tight_layout()
plt.show()
```
